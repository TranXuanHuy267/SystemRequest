```python
import openai
import re
import random
import numpy as np

# openai.api_key = "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"
# openai.api_key = "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"
openai.api_key = "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"
# openai.api_key = "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"

n = 1
temperature = 1
top_p = 1
num_of_gen_sent = 1

prompt_templates = [
    "Sử dụng tiếng Việt, sinh ra 5 câu văn ngắn, ra lệnh hoàn chỉnh có ý nghĩa mở một cái gì đó, chứa tất cả các từ sau đây: {}",
    "Sử dụng tiếng Việt, sinh ra 5 câu văn ngắn chứa nguyên văn tất cả các từ sau đây: {}",
]

def normalize_string(s):
    s = s.lower() # Convert the text to lowercase
    s = re.sub(' +', ' ', s) # Replace multiple spaces with a single space
    return s

def get_response(prompt, model_engine = "gpt-3.5-turbo", n=1, temperature=1.2, top_p=1):
    # model_engine='gpt-3.5-turbo-0613'
    response = openai.ChatCompletion.create(
        model=model_engine,
        n=n,
        messages=[
            {"role": "user", "content": prompt},
        ],
        temperature=temperature,
        top_p=top_p,
    )
    messages = [response.choices[i]['message'] for i in range(len(response.choices))]
    return messages[0]['content']

def get_prompt(prompt_templates, description, subdf, print_prompt=False):
    prompt = random.choice(prompt_templates).format(description + subdf)
    
    if print_prompt:
        print(f"Use prompt: \n{prompt}")
    return prompt

def get_answer(prompt_templates, description, subdf, print_prompt=False, n=1, temperature=1.2, top_p=1):
    prompt = get_prompt(prompt_templates, description, subdf, print_prompt)
    answer = get_response(prompt, n=n, temperature=temperature, top_p=top_p)
    return answer

def post_process_answer(answer):
    answer = answer.replace("\n", "")
    answer = re.sub(' +', ' ', answer)
    answer = answer.strip()
    return answer

def post_answer(answer):
    answer = answer[answer.find('.')+1:].strip()
    special_chas = [".", "!"]
    for cha in special_chas:
        while cha in answer:
            answer = answer.replace(cha, "")
    answer = answer.strip()
    answer = answer.lower()
    return answer
```

# From URL to output

```python
temp = final_dashboard[final_dashboard['URL']!='/pages/screen/6615']
d = {'URL': [], 'DONVI': [], 'CHUKY': []}
for url in list(set(temp['URL'])):
    d['URL'].append(url)
    short_donvi = 'a'*100
    for donvi in list(set(temp[temp['URL']==url]['donvi'])):
        if len(donvi) < len(short_donvi):
            short_donvi = donvi
    d['DONVI'].append(short_donvi)
    d['CHUKY'].append(list(set(temp[temp['URL']==url]['time']))[0])

url2output = pd.DataFrame(data=d)
```
