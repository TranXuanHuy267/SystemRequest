```python
import pandas as pd
from copy import deepcopy
import random
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
t = 1
speech_cmd = pd.read_excel("data/Cau lenh Speech Command.xlsx")
default_func = speech_cmd[4:]['Câu lệnh mẫu'].tolist()
default_func = [i.strip() for i in default_func]
default_func
```

```python
len(default_func)
```

```python
type_data = {
    'testcase': 'data_creation/processed_test_case.csv'
}
for type in type_data.keys():
    true_data = pd.read_csv(type_data[type])
    false_data = deepcopy(true_data)
    for i in range(len(default_func)):
        temp = deepcopy(default_func)
        temp.pop(i)
        random.shuffle(temp)
        indexes = false_data[false_data['Câu lệnh có sẵn']==default_func[i]].index
        
        for j in indexes:
            false_data['Câu lệnh có sẵn'].loc[j] = random.choice(temp)
    # false_data.drop(columns=['Unnamed: 0', 'Unnamed: 0.1'], inplace=True)
    false_data.to_csv("data_creation/test case/"+type+"_false.csv")
```
