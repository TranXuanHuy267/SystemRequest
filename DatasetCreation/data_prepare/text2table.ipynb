```python
import pandas as pd

dataset = pd.read_csv("/Users/tranxuanhuy/Documents/Viettel Digital Talent/Chatbot/Code/data_creation/generate_db_general.csv")
dataset.drop(columns=['Unnamed: 0.3', 'Unnamed: 0.2', 'Unnamed: 0.1', 'Unnamed: 0'], inplace=True)
def str2list(descript):
    sents = []
    while "\n\n" in descript:
        descript = descript.replace("\n\n", "\n")
    desc = descript.split('\n')
    for sent in desc:
        new_sent = sent
        if sent.find(".") != -1 and sent[sent.find(".")+1:].find(".") != -1:
            index1 = sent.find(".")+1
            index2 = sent[index1:].find(".")+index1
            new_sent = sent[index1+1:index2]
        elif sent.find(".") != -1 and sent[sent.find(".")-1] not in ['1', '2', '3', '4', '5']:
            index1 = sent.find(".")
            new_sent = sent[:index1]
        elif sent.find(".") != -1 and sent[sent.find(".")-1] in ['1', '2', '3', '4', '5']:
            index1 = sent.find(".")
            new_sent = sent[index1+2:]
        
        while "?" in new_sent:
            new_sent = new_sent.replace("?", "")
        sents.append(new_sent)
    return sents


dataset['processlist'] = dataset['Câu lệnh sinh ra'].apply(str2list)
profiles = ["đát bót", "đát bót chỉ huy",
            "đát bọt", "đát bọt chỉ huy",
            "dashboard", "dashboard chỉ huy",
            "db", "db chỉ huy"]

dataset['processlist'] = dataset['processlist'].apply(lambda lst: lst[:5] if len(lst)>5 else lst)
dataset['processlist'] = dataset['processlist'].apply(lambda lst: [lst[i] if ("dashboard" in lst[i] or "db" in lst[i]) else "" for i in range(len(lst))])
dataset['processlist'] = dataset['processlist'].apply(lambda lst: [item for item in lst if item != ""])
dataset['checklen'] = dataset['processlist'].apply(lambda lst: len([[i for i in profiles if i in inst] for inst in lst]))

while len(dataset['checklen'].value_counts().keys()) > 1:
    dataset['processlist'] = dataset['processlist'].apply(lambda lst: lst+[lst[-1]] if len(lst)<5 else lst)
    dataset['checklen'] = dataset['processlist'].apply(lambda lst: len([[i for i in profiles if i in inst] for inst in lst]))
```

```python
import random

def db_type_extract(sent):
    if "dashboard chỉ huy" in sent:
        return "dashboard chỉ huy"
    elif "db" in sent and "dbn" not in sent:
        return "db"
    elif "dashboard" in sent:
        return "dashboard"

def from_all_to_dashboard(sent, typee):
    sent = sent.replace(typee, "dashboard")
    if "  " in sent:
        sent = sent.replace("  ", " ")
    return sent

def from_dashboard_to_each(sent):
    typee = random.choice(list(typeOutput2Input.keys()))
    return typee

def from_each_to_specific(loai_bieu_do):
    typee = random.choice(list(typeOutput2Input[loai_bieu_do])).lower()
    return typee

def from_specific_to_sent(sent, loai_bieu_do):
    sent = sent.replace("dashboard", loai_bieu_do)
    if "  " in sent:
        sent = sent.replace("  ", " ")
    return sent
```

```python
# Create a new DataFrame by transforming the rows
new_rows = []
for index, row in dataset.iterrows():
    for name in row['processlist']:
        new_rows.append({'ID': row['ID'], 'câu lệnh sinh ra': name, 'đơn vị': row['donvi'], 'chu kỳ thời gian': row['time'], 'đơn vị (viết tắt)': url2shortestdonvi[row['URL']], 'URL': row['URL']})
prompt_dataset = pd.DataFrame(new_rows)
prompt_dataset['loại biểu đồ (raw)'] = prompt_dataset['câu lệnh sinh ra'].apply(db_type_extract)
prompt_dataset['câu lệnh sinh ra'] = prompt_dataset[['câu lệnh sinh ra', 'loại biểu đồ (raw)']].apply(lambda x: from_all_to_dashboard(x[0], x[1]), axis=1)
prompt_dataset['loại biểu đồ'] = prompt_dataset['câu lệnh sinh ra'].apply(from_dashboard_to_each)
prompt_dataset['loại biểu đồ (chi tiết)'] = prompt_dataset['loại biểu đồ'].apply(from_each_to_specific)
prompt_dataset['loại biểu đồ'] = prompt_dataset['loại biểu đồ'].str.lower()
prompt_dataset['câu lệnh sinh ra'] = prompt_dataset[['câu lệnh sinh ra', 'loại biểu đồ (chi tiết)']].apply(lambda x: from_specific_to_sent(x[0], x[1]), axis=1)
```

```python
def sorted_by_key(diction):
    key_sorted = [(len(key), key) for key in diction.keys()]
    key_sorted = sorted(key_sorted, reverse=True)
    diction = {j:diction[j] for i, j in key_sorted}
    return diction
```

```python
def remove_donvi(title_chart):
    features = title_chart.split("_")
    if " " not in features[0] and features[0] not in ["company", 'info', 'test', 'rank', 'rada', 'table', 'tooltip', '21', 'dgsdf', 'khdn', 'dt-cp-ln', 'arpu', 'mou', 'dou']:
        output =  features[0]
        output2 = "_".join(features[1:])
    elif features[0] in ['công ty mẹ', 'tập đoàn', 'ban cntt', 'khối tỉnh/tp', 'công đoàn', 'viettel tỉnh/tp', 'viettel tỉnh tp']:
        output = features[0]
        output2 = "_".join(features[1:])
    elif len(features) == 4 and " " not in features[2]:
        output = features[2]
        output2 = "_".join(features[:2]+features[3:])
    else:
        output = ""
        output2 = "_".join(features[0:])
    return output, output2
    
def remove_time(title_chart_rm_donvi):
    features = title_chart_rm_donvi.split("_")
    if len(features) == 1:
        features = "_".join(features)
        return features, ""
    output2 = ""
    output = deepcopy(features)
    for feature in features:
        for i in ['ngay', 'thang', 'nam', 'quy', 'ngày', 'tháng', 'năm', 'quý', 'tết', 'qúy']:
            if i in feature:
                output2 = feature
                output.remove(feature)
                break
    
    output = "_".join(output)
    return output, output2

def remove_redundant(title_chart_re_donvi_time):
    features = title_chart_re_donvi_time.split("_")
    output2 = ""
    output = deepcopy(features)
    if len(features) > 1:
        for feature in features:
            feature2 = feature.strip()
            if feature2 in ['line', 
                           'theo hướng', 
                           'xu huong', 
                           'theo cơ cấu', 
                           'cơ cấu từng khối', 
                           'cơ cấu cơ cấu từng khối', 
                           'cơ cấu dịch vụ', 
                           'cơ cấu đơn vị (lk kệ thống)',
                           'cơ cấu khối',
                           'các đơn vị',
                           'cơ cấu đơn vị',
                           'từng đơn vị',
                           'các đơn vị con nội bộ',
                           'cơ cấu %',
                           'theo loại thuê bao',
                           ] \
                or "theo hạ tầng" in feature:
                output2 = feature2
                output.remove(feature)
                break

            if 'test' in feature:
                output2 = feature
                output.remove(feature)
    output = "_".join(output)
    return output, output2

def check_final_name(row):
    if row['TITLE_CHART_RM_DONVI_TIME_REDUNDANT'] in row['TYPECHART']:
        return ""
    
    if row['TITLE_CHART_RM_DONVI_TIME_REDUNDANT'].isnumeric():
        return ""
    
    list_check2 = ['infomain', 'info_main', 'check', 'làm tròn', 'thoa', 'hà', 'copy', 'mai']
    for item in list_check2:
        if item in row['TITLE_CHART_RM_DONVI_TIME_REDUNDANT']:
            return ""

    if row['TITLE_CHART_RM_DONVI_TIME_REDUNDANT'][0].isnumeric() \
        and row['TITLE_CHART_RM_DONVI_TIME_REDUNDANT'][1] in [' ', '.']:
        output = row['TITLE_CHART_RM_DONVI_TIME_REDUNDANT'][2:].strip()
    else:
        output = row['TITLE_CHART_RM_DONVI_TIME_REDUNDANT'].strip()

    return output

def check_final_name_for_input(output):
    if len(output.split("_")) > 1:
        output = " ".join([i.strip() for i in output.split("_")])
    return output
```

```python
def add_chitieuInput(row):
    if row['loại biểu đồ'] == 'dashboard':
        chitieuTieudeInput = random.choice(list(chitieuTieudeOutput2inputDashboard.keys()))
        chitieuInput = chitieuTieudeOutput2inputDashboard[chitieuTieudeInput]
        chitieu = chitieuTieudeInput.split("....")[0]
        tieude = chitieuTieudeInput.split("....")[1]
    elif row['loại biểu đồ'] != 'khác':
        chitieuTieudeInput = random.choice(list(chitieuTieudeOutput2inputChart.keys()))
        chitieuInput = chitieuTieudeOutput2inputChart[chitieuTieudeInput]
        chitieu = chitieuTieudeInput.split("....")[0]
        tieude = chitieuTieudeInput.split("....")[1]
    else:
        chitieu = ""
        tieude = ""
        chitieuInput = ""
    
    if chitieu == "":
        chitieu='khác'
    if tieude == "":
        tieude = 'khác'
        
    return pd.Series([chitieuInput, chitieu, tieude], index=['chi tiêu input', 'chỉ tiêu', 'tiêu đề'])

def chitieu_position(row):
    chitieu_position = ['sau biểu đồ', 'trước đơn vị', 'sau đơn vị', 'trước thời gian', 'sau thời gian']
    return random.choice(chitieu_position)

def add_c_after_b(a, b, c):
    index = a.find(b)
    if index != -1:
        result = a[:index + len(b)] + " " + c + a[index + len(b):]
        return result.strip()
    else:
        return a
    
def add_c_before_b(a, b, c):
    index = a.find(b)
    if index != -1:
        result = a[:index] + c + " " + a[index:]
        return result.strip()
    else:
        return a

def add_chitieu_by_position(row):
    if row['vị trí chỉ tiêu'] == 'sau biểu đồ':
        output = add_c_after_b(row['câu lệnh sinh ra (mới)'], row['loại biểu đồ (chi tiết)'], row['chi tiêu input'])
    elif row['vị trí chỉ tiêu'] == 'trước đơn vị':
        output = add_c_before_b(row['câu lệnh sinh ra (mới)'], row['đơn vị mới (chi tiết)'], row['chi tiêu input'])
    elif row['vị trí chỉ tiêu'] == 'sau đơn vị':
        output = add_c_after_b(row['câu lệnh sinh ra (mới)'], row['đơn vị mới (chi tiết)'], row['chi tiêu input'])
    elif row['vị trí chỉ tiêu'] == 'trước thời gian':
        output = add_c_before_b(row['câu lệnh sinh ra (mới)'], row['thời gian (mới hẳn)'], row['chi tiêu input'])
    elif row['vị trí chỉ tiêu'] == 'sau thời gian':
        output = add_c_after_b(row['câu lệnh sinh ra (mới)'], row['thời gian (mới hẳn)'], row['chi tiêu input'])
    
    return output
    
        
```
