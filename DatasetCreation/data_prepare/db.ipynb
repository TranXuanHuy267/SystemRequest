```python
import pandas as pd

dataset = pd.read_csv("/Users/tranxuanhuy/Documents/Viettel Digital Talent/Chatbot/Code/data_creation/generate_db_general.csv")
dataset.drop(columns=['Unnamed: 0.3', 'Unnamed: 0.2', 'Unnamed: 0.1', 'Unnamed: 0'], inplace=True)
def str2list(descript):
    sents = []
    while "\n\n" in descript:
        descript = descript.replace("\n\n", "\n")
    desc = descript.split('\n')
    for sent in desc:
        new_sent = sent
        if sent.find(".") != -1 and sent[sent.find(".")+1:].find(".") != -1:
            index1 = sent.find(".")+1
            index2 = sent[index1:].find(".")+index1
            new_sent = sent[index1+1:index2]
        elif sent.find(".") != -1 and sent[sent.find(".")-1] not in ['1', '2', '3', '4', '5']:
            index1 = sent.find(".")
            new_sent = sent[:index1]
        elif sent.find(".") != -1 and sent[sent.find(".")-1] in ['1', '2', '3', '4', '5']:
            index1 = sent.find(".")
            new_sent = sent[index1+2:]
        
        while "?" in new_sent:
            new_sent = new_sent.replace("?", "")
        sents.append(new_sent)
    return sents


dataset['processlist'] = dataset['Câu lệnh sinh ra'].apply(str2list)
profiles = ["đát bót", "đát bót chỉ huy",
            "đát bọt", "đát bọt chỉ huy",
            "dashboard", "dashboard chỉ huy",
            "db", "db chỉ huy"]

dataset['processlist'] = dataset['processlist'].apply(lambda lst: lst[:5] if len(lst)>5 else lst)
dataset['processlist'] = dataset['processlist'].apply(lambda lst: [lst[i] if ("dashboard" in lst[i] or "db" in lst[i]) else "" for i in range(len(lst))])
dataset['processlist'] = dataset['processlist'].apply(lambda lst: [item for item in lst if item != ""])
dataset['checklen'] = dataset['processlist'].apply(lambda lst: len([[i for i in profiles if i in inst] for inst in lst]))

while len(dataset['checklen'].value_counts().keys()) > 1:
    dataset['processlist'] = dataset['processlist'].apply(lambda lst: lst+[lst[-1]] if len(lst)<5 else lst)
    dataset['checklen'] = dataset['processlist'].apply(lambda lst: len([[i for i in profiles if i in inst] for inst in lst]))
```

```python
import random

def db_type_extract(sent):
    if "dashboard chỉ huy" in sent:
        return "dashboard chỉ huy"
    elif "db" in sent and "dbn" not in sent:
        return "db"
    elif "dashboard" in sent:
        profiles = ["đát bót", "đát bót chỉ huy",
            "đát bọt", "đát bọt chỉ huy",
            "dashboard", "dashboard chỉ huy",
            "db", "db chỉ huy"]
        extends = [bieudo.lower() for bieudo in keyword['Tên biểu đồ']]
        khac_types = ["biểu đồ"] + [""]*6
        return random.choice(profiles+extends+khac_types)

def replace_db_type(sent, db_type):
    if "dashboard" in sent and "dashboard chỉ huy" not in sent:
        sent = sent.replace("dashboard", db_type)
    if "  " in sent:
        sent = sent.replace("  ", " ")
    return sent

def discriminate_db_type(db_type):
    profiles = ["đát bót", "đát bót chỉ huy",
            "đát bọt", "đát bọt chỉ huy",
            "dashboard", "dashboard chỉ huy",
            "db", "db chỉ huy"]
    extends = [bieudo.lower() for bieudo in keyword['Tên biểu đồ']]
    
    if db_type in profiles:
        return "dashboard"
    elif db_type in extends:
        return "biểu đồ"
    else:
        return "khác"    
```

```python
# Create a new DataFrame by transforming the rows
new_rows = []
for index, row in dataset.iterrows():
    for name in row['processlist']:
        new_rows.append({'ID': row['ID'], 'câu lệnh sinh ra': name, 'đơn vị': row['donvi'], 'chu kỳ thời gian': row['time'], 'đơn vị (viết tắt)': url2shortestdonvi[row['URL']], 'URL': row['URL']})
prompt_dataset = pd.DataFrame(new_rows)
prompt_dataset['loại biểu đồ (chi tiết)'] = prompt_dataset['câu lệnh sinh ra'].apply(db_type_extract)
prompt_dataset['câu lệnh sinh ra'] = prompt_dataset[['câu lệnh sinh ra', 'loại biểu đồ (chi tiết)']].apply(lambda x: replace_db_type(x[0], x[1]), axis=1)
prompt_dataset['loại biểu đồ'] = prompt_dataset['loại biểu đồ (chi tiết)'].apply(discriminate_db_type)
```
