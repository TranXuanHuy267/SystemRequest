```python
# Tên lệnh
lenhs = ["Mở", "mở"]

# Check: all profile name are in above profiles list
dashboard_cmd = dashboard['CODE']
dashboard['lenh'] = ""
dashboard['cmd_rm_lenh'] = ""
IDS = [i for i in range(5433)]
for i, line in enumerate(dashboard_cmd):
    count = 0
    for lenh in lenhs:
        if lenh in line:
            count += 1
            dashboard.loc[i, "lenh"] = lenh
            cmd_rm_lenh = line.replace(lenh, "")
            dashboard.loc[i, "cmd_rm_lenh"] = cmd_rm_lenh
            break
    if count > 0:
        IDS.remove(i)
print("List dashboard's id that are not have profile name in above 'cmds' list:")
print(IDS)
```

```python
# Tên profile
profiles = ["đát bót", "đát bọt", "dashboard", "Dashboard", "DB", "db"]

# Check: all profile name are in above profiles list
dashboard_cmd = dashboard['cmd_rm_lenh']
IDS = [i for i in range(5433)]
IDS.remove(568)
for i, line in enumerate(dashboard_cmd):
    count = 0
    for profile in profiles:
        if profile in line:
            count += 1
    if count > 0:
        IDS.remove(i)
        
print("List dashboard's id that are not have profile name in above 'profiles' list:")
print(IDS)

# Check 
```

```python
# Tên profile
profiles = ["đát bót", "đát bót chỉ huy",
            "đát bọt", "đát bọt chỉ huy",
            "dashboard", "dashboard chỉ huy",
            "Dashboard", "Dashboard chỉ huy",
            "DB", "DB chỉ huy",
            "db", "db chỉ huy"]

# Check: all profile name are in above profiles list
dashboard_cmd = dashboard['cmd_rm_lenh']
dashboard['profile'] = ""
dashboard['cmd_rm_lenh_profiles'] = ""

IDS = [i for i in range(5433)]
IDS.remove(568)
for i, line in enumerate(dashboard_cmd):
    count = 0
    fin = 0
    while fin == 0:
        fin = 1
        for profile in profiles[::-1]:
            if count == 0:
                if profile in line:
                    count += 1
                    fin = 0
                    dashboard.loc[i, "profile"] = profile
                    cmd_rm_profile = line.replace(profile, "")
                    dashboard.loc[i, "cmd_rm_lenh_profiles"] = cmd_rm_profile
                    break
            else:
                line = dashboard.loc[i, "cmd_rm_lenh_profiles"]
                if profile in line:
                    count += 1
                    fin = 0
                    # dashboard.loc[i, "profile"] = profile
                    cmd_rm_profile = line.replace(profile, "")
                    dashboard.loc[i, "cmd_rm_lenh_profiles"] = cmd_rm_profile
                    break

    if count > 0:
        IDS.remove(i)
        

print("List dashboard's id that are not have profile name in above 'profiles' list:")
print(IDS)
```

```python
# Thời gian & chu kỳ thời gian
db_times = ["ngày", "tháng", "năm", "Tháng"]
db_times = ["ngày", "tháng", 'Tháng']

# Check: all profile name are in above profiles list
dashboard_cmd = dashboard['cmd_rm_lenh_profiles']
dashboard['time'] = ""
dashboard['cmd_rm_lenh_profiles_times'] = ""
IDS = [i for i in range(5433)]
IDS.remove(568)
for i, line in enumerate(dashboard_cmd):
    count = 0
    for time in db_times:
        if time in line:
            count += 1
            dashboard.loc[i, "time"] = time
            cmd_rm_profile_time = line.replace(time, "")
            dashboard.loc[i, "cmd_rm_lenh_profiles_times"] = cmd_rm_profile_time
            break
    if count > 0:
        IDS.remove(i)
print("List dashboard's id that are not have profile name in above 'db_times' list:")
print(IDS)
```
