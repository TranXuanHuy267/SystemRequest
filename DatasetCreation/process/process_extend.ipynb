## Check đơn vị dashboard bởi Dashboard Type == 1

```python
dashboard_id = pd.read_csv("/Users/tranxuanhuy/Documents/Viettel Digital Talent/Chatbot/Code/data/CONFIG_DASHBOARD_202309051338.csv")
# dashboard_id.drop(columns=['UPDATE_TIME', 'UPDATE_USER', 'DESCRIPTION', 'STATUS'], inplace=True)
# dashboard_id = dashboard_id.applymap(lambda x: x.lower() if isinstance(x, str) else x)
dashboard_id['check_'] = dashboard_id['DASHBOARD_NAME'].apply(lambda x: 1 if x.find("_")>-1 else 0)
dashboard_id['check_2'] = dashboard_id.apply(lambda x: 1 if (x['check_']==1 and x['DASHBOARD_NAME'][x['DASHBOARD_NAME'].find("_")+1:]==str(x["ID"])) else 0, axis=1)
# dashboard_id = dashboard_id[dashboard_id['check_2']==0]
dashboard_id['DASHBOARD_NAME_NEW'] = dashboard_id.apply(lambda x: x['DASHBOARD_NAME'].replace("_"+str(x['ID']), "") if x['check_2']==1 else x['DASHBOARD_NAME'], axis=1)
condition = (dashboard_id['DASHBOARD_TYPE'] == 0) & (dashboard_id['MENU_ITEM_ID'].isna())
indices_to_drop = dashboard_id[condition].index
dashboard_id.drop(indices_to_drop, inplace=True)

profileid2donvi = {}
for profile_id in dashboard_id['PROFILE_ID'].unique():
    indexss = dashboard_id[(dashboard_id['PROFILE_ID']==profile_id)&(dashboard_id['DASHBOARD_TYPE']==1)].index
    if len(indexss) > 0:
        profileid2donvi[profile_id] = dashboard_id['DASHBOARD_NAME'].loc[dashboard_id[(dashboard_id['PROFILE_ID']==profile_id)&(dashboard_id['DASHBOARD_TYPE']==1)].index[0]]

dashboard_id['donvi'] = dashboard_id['PROFILE_ID'].apply(lambda x: profileid2donvi[x] if x in profileid2donvi.keys() else float('NaN'))
dashboard_id
```
