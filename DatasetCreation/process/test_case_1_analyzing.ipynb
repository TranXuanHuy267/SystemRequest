```python
speech_cmd = pd.read_excel("data/Cau lenh Speech Command.xlsx")
default_func = speech_cmd[4:]['Câu lệnh mẫu'].tolist()
default_func = [i.strip() for i in default_func]
default_func
```

```python
data = pd.read_excel("data/Testcase_Mở báo cáo_Chatbox.xlsx")
data.drop(columns=["Unnamed: 0", 'Unnamed: 5', 'Unnamed: 6', "Unnamed: 7", 'Unnamed: 8', "KỊCH BẢN KIỂM THỬ *"], inplace=True)
data.rename(columns={
    "Unnamed: 1": "Câu lệnh nhập vào",
    "Unnamed: 3": "Câu lệnh có sẵn",
    "Unnamed: 4": "Output",
    "Unnamed: 9": "Dự đoán"}, inplace=True
)
data = data[10:]
data.reset_index(inplace=True)
data.drop(columns=["index"], inplace=True)
special_index = data[data['Câu lệnh có sẵn'].isnull()].index.tolist()
data["Câu lệnh nhập vào"]=data["Câu lệnh nhập vào"].apply(lambda x: x[x.find('"')+1: x.find('"')+x[x.find('"')+1:].find('"')+1].strip() if type(x)!=float else "")
list_test_func = data['Câu lệnh nhập vào'].loc[special_index].tolist()
special_index.append(len(data))
special_index = pd.Int64Index(special_index)
cau_lenh_co_san = []
for i in range(len(data)):
    if i in special_index:
        cau_lenh_co_san.append("")
        continue
    for j in range(len(special_index)):
        if i > special_index[j] and i < special_index[j+1]:
            cau_lenh_co_san.append(list_test_func[j])
            break
    
data["Câu lệnh có sẵn"] = cau_lenh_co_san
data = data.loc[[i for i in range(len(data)) if i not in special_index]]
data.reset_index(inplace=True)
data.drop(columns=["index"], inplace=True)
data["Output"]=data["Output"].apply(lambda x: 1 if x=='P' else 0)
data["Dự đoán"]=data["Dự đoán"].apply(lambda x: x[x.find('"')+1: x.find('"')+x[x.find('"')+1:].find('"')+1].strip() if type(x)!=float else "")
data['Dự đoán'][data['Output']==1]=data['Câu lệnh có sẵn'][data['Output']==1]
data
```

```python
data2 = pd.read_csv("data_creation/generate_cmd_3.csv")
data2.drop(columns=['Unnamed: 0.2', 'Unnamed: 0.1', 'Unnamed: 0'], inplace=True)
data2
```

```python
dictionary = {
    'báo cáo': ['bc', 'báo cáo', 'báo cáo'],
    'công thức': ['ct', 'công thức', 'công thức', 'công thức', 'công thức'],
    'cảnh báo': ['cb', 'cảnh báo', 'cảnh báo'],
    'kinh doanh': ['kd', 'kinh doanh', 'kinh doanh', 'kinh doanh', 'kinh doanh'],
    'dữ liệu': ['dl', 'dữ liệu', 'dữ liệu'],
    'chấm điểm': ['cđ', 'chấm điểm', 'chấm điểm', 'chấm điểm', 'chấm điểm'],
    'tài chính': ['tc', 'tài chính', 'tài chính', 'tài chính', 'tài chính'],
    'quản lý': ['ql', 'quản lý', 'quản lý'],
    'hướng dẫn': ['hd', 'hướng dẫn', 'hướng dẫn', 'hướng dẫn', 'hướng dẫn'],
    'chỉ tiêu': ['ct', 'chỉ tiêu', 'chỉ tiêu', 'chỉ tiêu', 'chỉ tiêu'],
    'kết quả': ['kq', 'kết quả', 'kết quả'],
    'phát triển': ['pt', 'phát triển', 'phát triển', 'phát triển', 'phát triển'],
    
}
```

```python
data2['Câu lệnh sinh ra'] = data2['Câu lệnh sinh ra'].apply(lambda x: x.lower())
list_new_cmd = []
for cmd in data2['Câu lệnh sinh ra']:
    new_cmd = cmd
    for key in dictionary.keys():
        if key in cmd:
            new_cmd = new_cmd.replace(key, random.choice(dictionary[key]))
    list_new_cmd.append(new_cmd)

data2['Câu lệnh sinh ra'] = list_new_cmd
```
